```python
# Two Sum: Given an array of integers, return indices of the two numbers such that they add up to a specific target.
def two_sum(target_value, given_array):
    val_result = []
    idx_result = []
    for i, a in enumerate(given_array):
        b = target_value - a
        search_array = given_array[min(i+1, len(given_array)-1):]
        if b in search_array:
            j = given_array.index(b)
            val_result.append((a, b))
            idx_result.append((i, j))
            given_array.remove(a)
            given_array.remove(b)

    return val_result, idx_result

given_array = [random.randint(0, 9) for _ in range(10)]
target_value = 10
val_result, idx_result = two_sum(target_value, given_array)
print(val_result)
print(idx_result)
```

```python
# Remove Duplicates from Sorted Array: Given a sorted array, remove the duplicates in-place such that each element appears only once and returns the new length.

def remove_duplicates(sorted_array):

    unique_vals = []
    for num in sorted_array:
        if num not in unique_vals:
            unique_vals.append(num)

    return unique_vals

given_array = [random.randint(0, 9) for _ in range(20)]
print(f"Given array: {given_array}")
sorted_array = sorted(given_array)
print(f"Unique array: {remove_duplicates(sorted_array)}")
```

```python
# Maximum Subarray: Find the contiguous subarray (containing at least one number) which has the largest sum and return its sum 
# (start from the first element, not anywhere in the array)

def maximum_subarray(given_array):
 
    if not given_array:
        return 0

    # Initialize current sum and maximum sum
    current_sum = max_sum = given_array[0]

    # Iterate through the array starting from the second element
    for num in given_array[1:]:
        # Update current sum
        current_sum = max(num, current_sum + num)
        # Update maximum sum
        max_sum = max(max_sum, current_sum)

    return max_sum

given_array = [random.randint(-9, 9) for _ in range(40)]
print(f"Given array: {given_array}")
print(f"Maximum subarray: {maximum_subarray(given_array)}")
```

```python
# Climbing Stairs: Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?

def climbing_stairs(num_stairs):

    n_double_types = 1+int(0.5 * num_stairs)
    n_cases = 0

    for n_double_combined_stairs in range(n_double_types):
        if num_stairs - n_double_combined_stairs >= n_double_combined_stairs:
            n_cases += math.comb(num_stairs - n_double_combined_stairs, n_double_combined_stairs)

    return n_cases

# Much better answer by dynamic program
def climb_stairs(n):
    # Base cases
    if n <= 1:
        return 1
    if n == 2:
        return 2
    
    # Initialize the base values for dynamic programming
    one_step_before = 2
    two_steps_before = 1
    
    # Calculate the number of ways for each step
    for i in range(3, n + 1):
        # The number of ways to reach the current step is the sum
        # of the ways to reach the one step before and two steps before
        current = one_step_before + two_steps_before
        # Update the steps
        two_steps_before = one_step_before
        one_step_before = current
    
    # The final value of one_step_before is the answer
    return one_step_before


num_stairs = 4
print(climbing_stairs(num_stairs))
print(climb_stairs(num_stairs))
```

```python
# Plus One: Given a non-empty array of digits representing a non-negative integer, increment one to the integer.

def plus_one(given_array):

    for t in range(len(given_array)-1, -1, -1):
        if given_array[t] < 9:
            given_array[t] += 1
            return given_array
        else:
            given_array[t] = 0

    return [1] + given_array

print(plus_one([9, 9, 9]))
```

```python
# Valid Parentheses: Given a string containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

def validate_parantheses(text):
    
    special_chars = ['(', ')', '{', '}', '[', ']']
    count_dic = {key: 0 for key in special_chars}

    for t in text:
        count_dic[t] += 1

        if count_dic['}'] > count_dic['{'] or count_dic[')'] > count_dic['('] or count_dic[']'] > count_dic['[']:
            return False
        
    return True

print(validate_parantheses('{{(()))}}'))
```

```python
# Best Time to Buy and Sell Stock: Find the maximum profit you can achieve. You may not engage in multiple transactions at the same time.

def max_profit_multi_trade(given_array):

    profit = 0

    if len(given_array) < 2:
        print("The array length should be greater than 1!")
        return profit

    for t, value in enumerate(given_array):

        if t == 0:
            buy_value = value
        else:
            profit += max(0, value - buy_value)
            buy_value = value
    
    return profit

def max_profit_single_trade(prices):
    if not prices:
        return 0

    max_profit = 0
    min_price = prices[0]

    for price in prices:
        min_price = min(min_price, price)
        profit = price - min_price
        max_profit = max(max_profit, profit)

    return max_profit

given_array = [random.randint(0, 9) for _ in range(10)]
print(f"Given array: {given_array}")
print(f"Maximum profit in Multi-trade mode: {max_profit_multi_trade(given_array)}")
print(f"Maximum profit in Single-trade mode: {max_profit_single_trade(given_array)}")
        
```

```python
# Single Number: Given a non-empty array of integers, every element appears twice except for one. Find that single one.

from collections import Counter

def find_single(given_array):

    counter_dic = Counter(given_array)
    for k in counter_dic.keys():
        if counter_dic[k] == 1:
            return k
    
    print('The input is not valid!')
    return None

given_array = [5, 5, 9, 9, 8, 2, 8]
print(f"Given array: {given_array}")
print(find_single(given_array))
```

```python
# Longest Substring Without Repeating Characters: Find the length of the longest substring without repeating characters.

def longest_unrepeated_substring(text):

    if len(text) == 0:
        return text
    
    answer = ''
    unique_text = []
    
    for t in text:

        if t not in unique_text:
            unique_text.append(t)
        else:
            if len(unique_text) > len(answer):
                answer = ''.join(unique_text)
            unique_text = []

    return answer

text = 'nimalovesmina'
print(longest_unrepeated_substring(text))        
```

```python
# Container With Most Water: Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it can trap after raining.

def water_container(height):
    
    if not height:
        return 0

    trapped_water = 0
    left, right = 0, len(height) - 1
    left_max, right_max = height[left], height[right]

    while left < right:
        print(f"left: {left}")
        print(f"height[left]: {height[left]}")
        print(f"left_max: {left_max}")
        print(f"right: {right}")
        print(f"height[right]: {height[right]}")
        print(f"right_max: {right_max}")

        if height[left] < height[right]:
            if height[left] > left_max:
                left_max = height[left]
            else:
                trapped_water += left_max - height[left]
            left += 1
        else:
            if height[right] > right_max:
                right_max = height[right]
            else:
                trapped_water += right_max - height[right]
            right -= 1

        print(trapped_water)

    return trapped_water

given_array = [5, 9, 9, 2, 2, 7, 0, 8, 8, 1]
print(given_array)
print(water_container(given_array))
```

```python
# Group Anagrams: Given an array of strings, group anagrams together.

def group_anagrams(given_array):

    my_dic = {}

    for word in given_array:
        
        potential_key = tuple(sorted(word))
        if potential_key not in my_dic.keys():
            my_dic[potential_key] = [word]
        else:
            my_dic[potential_key].append(word)

    output = []
    for value in my_dic.values():
        output.append(value)

    return output


given_array = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(group_anagrams(given_array))
```

```python
# Combination Sum II: Given a collection of candidate numbers and a target number, find all unique combinations in candidates where the candidate numbers sum to the target.

def combination_sum2(candidates, target):
    
    def backtrack(start, path, target, back_idx):

        print('=================================')
        print(f'--------- back_idx is {back_idx}')
        print(f'start is {start}')
        print(f'target is {target}')
        print(f'path is {path}')

        if target == 0:
            result.append(path)
            return
        if target < 0:
            return

        for i in range(start, len(candidates)):

            # Skip duplicates
            if i > start and candidates[i] == candidates[i - 1]:
                continue

            new_start = i + 1
            new_path = path + [candidates[i]]
            new_target = target - candidates[i]
            new_back_idx = back_idx + 1

            # Include the candidate number in the path
            backtrack(new_start, new_path, new_target, new_back_idx)
    
    candidates.sort()  # Sort the candidates to handle duplicates
    result = []

    back_idx = 0
    backtrack(0, [], target, back_idx)
    return result

# Example usage
candidates = [10, 1, 2, 7, 6, 1, 5]
target = 8
print(combination_sum2(candidates, target))
```

```python
# Letter Case Permutation: Given a string s, we can transform every letter individually to be lowercase or uppercase to create another string. Return a list of all possible strings we could create.

def letter_case_permutation(text):
    
    def backtrack(path, index):

        if len(path) == len(text):
            result.append(path)
            return

        new_index = index + 1
        if text[index].isalpha():
            new_path = path + text[index].lower()
            backtrack(new_path, new_index)
            new_path = path + text[index].upper()
            backtrack(new_path, new_index)

        else:
            new_path = path + text[index]
            backtrack(new_path, new_index)

        
    result = []
    path = ''
    index = 0
    backtrack(path, index)

    return result

# Example usage
text = 'a1b2'
print(letter_case_permutation(text))
```

```python
# Max Increase to Keep City Skyline: Given an n x n grid representing the heights of buildings, find the maximum total sum that the height of the buildings can be increased 
# without changing the city's skyline from any cardinal direction.

def maxIncreaseKeepingSkyline(grid):

    row_maxes = [max(row) for row in grid]
    print(row_maxes)
    col_maxes = [max(col) for col in zip(*grid)]
    print(col_maxes)

    res = 0
    for r, row in enumerate(grid):
        for c, val in enumerate(row):
            res += min(row_maxes[r], col_maxes[c]) - val
    
    return res

# Assuming 'grid' is your n x n grid
grid = [
    [1, 2, 3, 4], 
    [5, 4, 2, 4],
    [3, 6, 1, 5],
    [2, 3, 1, 5]
]
print(maxIncreaseKeepingSkyline(grid))
```

```python
# Text Justification: Given an array of words and a width maxWidth, format the text such that each line has exactly maxWidth characters.

def full_justify(words, maxWidth):
    result = []
    current_line = []
    current_length = 0

    for word in words:

        # Check if adding the next word would exceed maxWidth
        if current_length + len(word) + len(current_line) > maxWidth:

            # Time to justify the current line
            for i in range(maxWidth - current_length):
                current_line[i % (len(current_line) - 1 or 1)] += ' '

            result.append(''.join(current_line))
            current_line, current_length = [], 0

        # Add the current word to the line
        current_line.append(word)
        current_length += len(word)

    # Last line - left-justified
    result.append(' '.join(current_line).ljust(maxWidth))

    return result


# Example usage
words = ["This", "is", "an", "example", "of", "text", "justification."]
max_width = 16
justified_text = full_justify(words, max_width)
for line in justified_text:
    print(f'"{line}"')
```

```python
# Regular Expression Matching: Implement regular expression matching with support for '.' and '*'.

def isMatch(s, p):
    dp = [[False] * (len(p) + 1) for _ in range(len(s) + 1)]
    print(dp)
    dp[-1][-1] = True
    for i in range(len(s), -1, -1):
        for j in range(len(p) - 1, -1, -1):
            first_match = i < len(s) and p[j] in {s[i], '.'}
            if j+1 < len(p) and p[j+1] == '*':
                dp[i][j] = dp[i][j+2] or first_match and dp[i+1][j]
            else:
                dp[i][j] = first_match and dp[i+1][j+1]
    return dp[0][0]

s = "aab"
p = "c*a*b"
print(isMatch(s, p))
```

```python
# Largest Rectangle in Histogram: Given n non-negative integers representing the histogram's bar height where the width of each bar is 1, find the area of the largest rectangle in the histogram.

def largestRectangleArea(heights):
    heights.append(0)
    stack = [-1]
    ans = 0
    for i in range(len(heights)):
        print('===========================================')
        print(f"heights[i] = {heights[i]}")
        print(f"heights[stack[-1]] = {heights[stack[-1]]}")
        while heights[i] < heights[stack[-1]]:
            print('----------------------------')
            print(f"heights[i] = {heights[i]}")
            print(f"stack[-1]={stack[-1]}")
            print(f"heights[stack[-1]] = {heights[stack[-1]]}")
            h = heights[stack.pop()]
            print(f"h={h}")
            print(f"i={i}")
            print(f"stack[-1]={stack[-1]}")
            w = i - stack[-1] - 1
            print(f"w=i-stack[-1]-1={w}")
            ans = max(ans, h * w)
            print(f"ans=max(ans, h*w)={ans}")
        stack.append(i)
        print(f"stack = {stack}")
    heights.pop()
    print(f"heights = {heights}")
    return ans

print(largestRectangleArea([3, 6, 1, 1, 1, 1, 1, 1]))
```
